## Unificação de todas as lojas num só dataframe e tratamento de dados
Unificando todas as lojas num único dataframe já que tem a mesma estrutura e realizando pequenos ajustes técnicos para facilitar a análise.

```python
# Adicionando uma coluna 'Loja' em cada dataframe para identificar a origem dos dados após a unificação
loja['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Unificação dos dados concatenando os 4 dataframes em um único chamado 'df' e usando o parâmetro ignore_index=True para evitar indíces duplicados
df = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

# Verificação exibindo 5 amostras aleatórias para garantir que as lojas foram misturadas corretamente
display(df.sample(5))
```

```python
# Diagnóstico de qualidade para garantir que não há nenhum item nulo
print(' Quandidade de nulos ')
display(df.isnull().sum())

# Verificando se há itens duplicados
duplicados = df.duplicated().sum()
print(f'Linhas duplicads encontradas: {duplicados}')
```

```python
# Ajustando datas para que o pandas entenda que a coluna 'Data da Compra' é tempo, não texto e usando o dayfirst=True pra garantir o formato brasileiro que é Dia/Mês/Ano
df['Data da Compra'] = pd.to_datetime(df['Data da Compra'], dayfirst=True)

# Criando colunas temporais (para análises futuras) Isso permitirá avaliar melhor o faturamento em cada período
df['Ano'] = df['Data da Compra'].dt.year
df['Mes'] =  df['Data da Compra'].dt.month

# Arredondado valores financeiros
df[['Preço', 'Frete']] = df[['Preço', 'Frete']].round(2)

# Verificação final
display(df.sample(5))
```
