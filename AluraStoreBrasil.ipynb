## Unificação de todas as lojas num só dataframe e tratamento de dados
Unificando todas as lojas num único dataframe já que tem a mesma estrutura e realizando pequenos ajustes técnicos para facilitar a análise.

```python
# Adicionando uma coluna 'Loja' em cada dataframe para identificar a origem dos dados após a unificação
loja['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Unificação dos dados concatenando os 4 dataframes em um único chamado 'df' e usando o parâmetro ignore_index=True para evitar indíces duplicados
df = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

# Verificação exibindo 5 amostras aleatórias para garantir que as lojas foram misturadas corretamente
display(df.sample(5))
```

```python
# Diagnóstico de qualidade para garantir que não há nenhum item nulo
print(' Quandidade de nulos ')
display(df.isnull().sum())

# Verificando se há itens duplicados
duplicados = df.duplicated().sum()
print(f'Linhas duplicads encontradas: {duplicados}')
```

```python
# Ajustando datas para que o pandas entenda que a coluna 'Data da Compra' é tempo, não texto e usando o dayfirst=True pra garantir o formato brasileiro que é Dia/Mês/Ano
df['Data da Compra'] = pd.to_datetime(df['Data da Compra'], dayfirst=True)

# Criando colunas temporais (para análises futuras) Isso permitirá avaliar melhor o faturamento em cada período
df['Ano'] = df['Data da Compra'].dt.year
df['Mes'] =  df['Data da Compra'].dt.month

# Arredondado valores financeiros
df[['Preço', 'Frete']] = df[['Preço', 'Frete']].round(2)

# Verificação final
display(df.sample(5))
```

```python
## Importando as outras bibliotecas que serão usadas
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Análise do faturamento



```python
# Agrupando pelo identificador da Loja e somando o faturamento
faturamento_por_loja = df.groupby('Loja')['Preço'].sum().reset_index(name='faturamento_total').sort_values('faturamento_total', ascending=False)

# Exibindo o faturamento total
display(faturamento_por_loja.style.format({'faturamento_total': '{:,.2f}'}, thousands='.', decimal=','))

```

```python
# Criando gráfico visual com seaborn
plt.figure(figsize=(10,6))
grafico_fat = sns.barplot(x='Loja', y='faturamento_total', hue='Loja', data=faturamento_por_loja, palette='viridis', legend=False)
plt.title('Ranking de faturamento por loja (R$)', fontsize=16, fontweight='bold')
plt.ylabel('Faturamento (R$)', fontsize=14)
plt.xlabel('Loja', fontsize=14)

# Adicionando rótulos de dados nas barras
for i in grafico_fat.containers:
    grafico_fat.bar_label(i, fmt=lambda v: f'R$ {v/1_000_000:.2f}M',fontsize=12, color='black', padding=3)

# Usando a função tight_layout para ajustar o layout do gráfico
plt.tight_layout()

# Salvando o graáfico para o reademe
plt.savefig('grafico_faturamento.png')
plt.show()
```

# 2. Vendas por categoria

```python
# Contagem de vendas por categorias em cada loja usando unstack() para transformar as categorias em colunas pra facilitar a leitura visual
vendas_categoria = df.groupby(['Loja', 'Categoria do Produto']).size().unstack().fillna(0)

# Visualizando a tabela com destaque para as categorias mais populares de cada loja com fundo laranja
display(vendas_categoria.style.format({col: '{:,.0f}' for col in vendas_categoria.columns}).highlight_max(color='orange', axis=1))
```

```python
# Criando o gráfico de barras empilhadas para analisar melhor as vendas das categorias de cada loja
ax = vendas_categoria.plot(kind='bar', stacked=True, figsize=(12,7), colormap='tab10')
plt.title('Vendas por categoria de produto', fontsize=16, fontweight='bold')
plt.ylabel('Quantidade de vendas', fontsize=14)

# Obtendo os handles do gráfico
handles, _ = ax.get_legend_handles_labels()

# Usando as colunas do DataFrame como rótulos da legenda para garantir a ordem (invertendo ambas handles e labels para corresponder à ordem de empilhamento)
plt.legend(handles[::-1], vendas_categoria.columns[::-1], title='Categorias', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Salvando gráfico
plt.savefig('grafico_vendas_categorias.png')
plt.show()
```

```python
# Faturamento total de cada categoria
faturamento_por_categoria = df.groupby('Categoria do Produto')['Preço'].sum().reset_index(name='faturamento_total').sort_values('faturamento_total', ascending=False)

# Exibindo a tabel com o faturamento toal por cada categoria
display(faturamento_por_categoria.style.format(
    {'faturamento_total': 'R${:,.2f}'}, thousands='.', decimal=','
))
```

```python
# Porcentagem de faturamento total de cada categoria para analisar as que mais se destacam
porcentagem_faturamento_categoria = (faturamento_por_categoria['faturamento_total'] / faturamento_por_categoria['faturamento_total'].sum())

# Criando um DataFrame com as categorias e suas porcentagens
porcentagem_df = faturamento_por_categoria[['Categoria do Produto']].copy()
porcentagem_df['porcentagem_total'] = porcentagem_faturamento_categoria

# Ordenando por porcentagem em ordem ascendente
porcentagem_df = porcentagem_df.sort_values('porcentagem_total', ascending=False)

# Exibindo a porcentagem de faturamento total de cada categoria com formatação
display(porcentagem_df.style.format({'porcentagem_total': '{:.2%}'}))

```

```python
# Criando o gráfico de pizza
plt.figure(figsize=(10, 8))
plt.pie(porcentagem_df['porcentagem_total'], labels=None, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('tab20'))
plt.title('Distribuição Percentual de Faturamento por Categoria', fontsize=16, fontweight='bold')

# Adicionando legenda com as categorias em ordem ascendente de porcentagem
plt.legend(porcentagem_df['Categoria do Produto'], title="Categorias", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.axis('equal')

# Salvando gráfico de porcentagem
plt.savefig('grafico_porcentagem_faturamento.png')
plt.tight_layout()
plt.show()
```

# 3. Média de avaliação das lojas


```python
# Calculando a média das estrelas de avaliações de cada loja
avaliacao_media = df.groupby('Loja')['Avaliação da compra'].mean().reset_index(name='media_avaliacao').sort_values('media_avaliacao', ascending=False).round(2)

# Visualizando a média das avaliações de cada loja
display(avaliacao_media)
```

```python
# Agrupando por categoria para calcular a avaliação média
avaliacao_por_categoria = df.groupby('Categoria do Produto')['Avaliação da compra'].mean().reset_index().sort_values('Avaliação da compra', ascending=False).round(2)

# Exibindo a média das avaliações por categoria
display(avaliacao_por_categoria)
```

# 4. Produtos Mais e Menos Vendidos

```python
# Descobrindo os produtos mais e menos vendidos de cada loja - verificando os 5 mais vendidos e os 5 menos vendidos
produtos_mais_vendidos = df.groupby(['Loja', 'Produto']).size().reset_index(name='quantidade_vendida').sort_values('quantidade_vendida', ascending=False).head(5)
produtos_menos_vendidos = df.groupby(['Loja', 'Produto']).size().reset_index(name='quantidade_vendida').sort_values('quantidade_vendida', ascending=True).head(5)

# Exibindo o resultado com os 5 mais vendidos e os 5 menos vendidos de cada loja
display(produtos_mais_vendidos)
print('----------')
display(produtos_menos_vendidos)
```

```python
vendas_por_loja_produto = df.groupby(['Loja', 'Produto']).size().reset_index(name='quantidade_vendida')

for loja_nome in df['Loja'].unique():
    print(f'\n--- Loja: {loja_nome} ---')

    # Produtos mais vendidos
    mais_vendidos = vendas_por_loja_produto[vendas_por_loja_produto['Loja'] == loja_nome].sort_values('quantidade_vendida', ascending=False).head(5)

    # Produtos menos vendidos
    menos_vendidos = vendas_por_loja_produto[vendas_por_loja_produto['Loja'] == loja_nome].sort_values('quantidade_vendida', ascending=True).head(5)

    # Combinando os 5 mais vendidos e 5 menos vendidos em uma única tabela, sem a coluna 'Loja' e sem o índice
    print('5 produtos mais vendidos e 5 produtos menos vendidos:')
    display(pd.concat([mais_vendidos.drop(columns=['Loja']), menos_vendidos.drop(columns=['Loja'])]).style.hide(axis='index'))
```

```python
# Calculando a quantidade de vendas para cada produto por loja
vendas_por_loja_produto = df.groupby(['Loja', 'Produto', 'Categoria do Produto']).size().reset_index(name='quantidade_vendida')

# Preparando uma lista para armazenar os dados do resumo
dados_resumo = []

for nome_loja in df['Loja'].unique():
    # Filtrando os dados para a loja atual
    vendas_loja = vendas_por_loja_produto[vendas_por_loja_produto['Loja'] == nome_loja]

    # Obtendo o produto mais vendido para a loja atual
    mais_vendido = vendas_loja.sort_values(by='quantidade_vendida', ascending=False).iloc[0]

    # Obtendo o produto menos vendido para a loja atual
    menos_vendido = vendas_loja.sort_values(by='quantidade_vendida', ascending=True).iloc[0]

    # Adicionando dados à lista de resumo
    dados_resumo.append({
        'loja': nome_loja,
        'produto_mais_vendido': mais_vendido['Produto'],
        'quantidade_mais_vendido': mais_vendido['quantidade_vendida'],
        'categoria_mais_vendido': mais_vendido['Categoria do Produto'],
        'produto_menos_vendido': menos_vendido['Produto'],
        'quantidade_menos_vendido': menos_vendido['quantidade_vendida'],
        'categoria_menos_vendido': menos_vendido['Categoria do Produto']
    })

# Criando um DataFrame a partir dos dados do resumo
summary_df = pd.DataFrame(dados_resumo)

# Exibindo a tabela de resumo final
display(summary_df)
```

```python
total_vendas_por_produto = df.groupby('Produto').size().reset_index(name='quantidade_vendida')

print('5 Produtos Mais Vendidos (Todas as Lojas):')
produtos_mais_vendidos_geral = total_vendas_por_produto.sort_values('quantidade_vendida', ascending=False).head(5)
display(produtos_mais_vendidos_geral)

print('\n5 Produtos Menos Vendidos (Todas as Lojas):')
produtos_menos_vendidos_geral = total_vendas_por_produto.sort_values('quantidade_vendida', ascending=True).head(5)
display(produtos_menos_vendidos_geral)
```

# 5. Frete médio por loja

```python
# Calculando a média do frete de cada loja
frete_medio = df.groupby('Loja')['Frete'].mean().reset_index(name='Média do frete').sort_values('Média do frete', ascending=False).round(2)

# Exibindo a média do frete de cada loja com o formato R$
display(frete_medio.style.format({'Média do frete': 'R$ {:,.2f}'}, decimal=',', thousands='.'))
```

```python
# Criando gráfico simples de comparação
plt.figure(figsize=(8,8))
grafico_frete = sns.barplot(x='Loja', y='Média do frete', hue='Loja', data=frete_medio, palette='magma', legend=False)
plt.title('Custo médio do frete de cada loja', fontsize=16, fontweight='bold')
plt.ylabel('Valor médio')

# Adicionando rótulos de dados nas barras
for i in grafico_frete.containers:
    grafico_frete.bar_label(i, fmt=lambda v: f'R$ {v:,.0f}',fontsize=12, color='black', padding=3)

# Salvando o gráfico
plt.tight_layout
plt.savefig('grafico_frete.png')
plt.show()
```

# Gráficos de comparação e análise final

```python
# Calculando e visualizando a tendência de lucro mensal para cada loja, gerando um gráfico de linhas e fornecendo insights.

# Criamos uma nova coluna no DF chamada 'lucro' subtraindo o valor de frete do preço pra incluir nessa coluna
df['Lucro'] = df['Preço'] - df['Frete']

# Visualizando o df com a nova coluna de lucro
display(df.head())
```

```python
# Calculando o lucro mensal de cada loja com groupby
lucro_mensal_por_loja = df.groupby(['Loja', 'Ano', 'Mes'])['Lucro'].sum().reset_index()
lucro_mensal_por_loja = lucro_mensal_por_loja.sort_values(by=['Loja', 'Ano', 'Mes'])
display(lucro_mensal_por_loja.head())
```

```python
# Calculanod lucro mensal médio anual (pra melhor visualização)
lucro_mensal_medio_anual = lucro_mensal_por_loja.groupby(['Loja', 'Ano'])['Lucro'].mean().reset_index()
display(lucro_mensal_medio_anual)
```

```python
# Gerando gráfico pra analisar a tendência da média mensal de cada loja por ano
custom_palette = {
    'Loja 1': 'blue',
    'Loja 2': 'orange',
    'Loja 3': 'green',
    'Loja 4': 'red'
}

plt.figure(figsize=(10, 6))
sns.lineplot(data=lucro_mensal_medio_anual, x='Ano', y='Lucro', hue='Loja', marker='o', palette=custom_palette)

plt.title('Lucro Mensal Médio Anual por Loja', fontsize=16, fontweight='bold')
plt.xlabel('Ano', fontsize=14)
plt.ylabel('Lucro Médio (R$)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Loja')
plt.tight_layout()
plt.savefig('grafico_lucro_mensal_medio_anual.png')
plt.show()
```

```python
# Calculando a receita média de cada loja por venda
receita_media_por_loja = df.groupby('Loja')['Preço'].mean().reset_index(name='receita_media_por_venda')
display(receita_media_por_loja)
```

```python
# Criando gráfico com linha da receita média por venda de cada loja
plt.figure(figsize=(8,5))
receita_media_por_loja.plot(kind='line', marker='o', linestyle='--', color='orange')
plt.title('Receita média por venda de cada loja')
plt.ylabel('Receita média por venda')
plt.xlabel('Loja')
plt.grid(True)
for index, row in receita_media_por_loja.iterrows():
    plt.text(index, row['receita_media_por_venda'], f'R$ {row['receita_media_por_venda']:.2f}', color='black', ha="center", va='bottom')
plt.tight_layout()
plt.savefig('grafico_receita_media_por_loja.png', format='png', dpi=300, bbox_inches='tight')
plt.show()
```

# Relatório de consultoria estratégica: Otimização da rede Alura Store

Este relatório tem como objetivo analisar o desempenho das filiais da Alura Store para embasar uma decisão estratégica importante: identificar qual unidade apresenta o menor desempenho operacional e financeiro, justificando um possível fechamento.

Para construir essa análise, processei a base de dados utilizando a biblioteca **Pandas**, focando em KPIs essenciais para o negócio, como o faturamento total bruto, a rentabilidade por categorias (com destaque para eletrônicos, móveis e brinquedos), a eficiência logística (impacto do frete) e as métricas de satisfação do cliente. Para garantir que os dados contassem uma história clara, utilizei **Seaborn** e **Matplotlib** na geração das visualizações.

Os dados revelaram um cenário bem nítido de desempenho crítico. A **Loja 4** apresentou, de forma consistente, o faturamento e o lucro médio mensal mais baixos de toda a rede. Ao fazer uma avaliação de mercado mais aprofundada, notei que mesmo no setor de Eletrônicos — que é o carro-chefe da empresa — a Loja 4 não conseguiu converter volume em um lucro sustentável.

Outro ponto que chamou atenção foi a relação de custo logístico versus receita. Embora a Loja 4 possua o frete médio mais baixo, essa vantagem teórica não se traduziu em rentabilidade real, o que indica um problema claro de ticket médio baixo ou volume de vendas insuficiente.
